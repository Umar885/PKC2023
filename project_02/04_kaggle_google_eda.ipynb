# How to Dealing with Price column

```python
df['Price'].value_counts()
```

```python
# check this missing values in price column
df['Price'].isnull().sum()
```

Remove $ sign from all columns using pandas

```python
df['Price'] = df['Price'].str.replace('$','')
df['Price'] = pd.to_numeric(df['Price'])


df['Price'].value_counts()
```

```python
df.head()
```

```python
df['Size'] = df['Size'].str.replace('Varies with device','Nan')


df['Size'].value_counts()
```
