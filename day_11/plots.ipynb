```python
# import Laibreries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import dataset
kashti = sns.load_dataset('titanic')
kashti
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot(x="survived", hue="sex", data=kashti)
```

```python
# scater plot   
plt.scatter(data=kashti, x="age", y="fare")
```

```python
sns.scatterplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
sns.barplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins=10, figsize=(9,7), grid=False)
```

```python
sns.boxplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
sns.boxenplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
# Assignment: How to change the x and y axis titles?
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="pclass", margin_titles=True, palette={1:"seagreen", 0:"gray"})
g=g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend();
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="sex", margin_titles=True, palette={1:"seagreen", 0:"gray"})
g=g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend();
```

```python
sns.stripplot(data=kashti, x="sex", y="fare", hue="class")
```

```python
corr = kashti.corr()
```

```python
sns.violinplot(data=kashti, x="sex", y="fare", hue="class")
```
