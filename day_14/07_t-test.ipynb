```python
# import libraries

import seaborn as sns
import pandas as pd
from scipy.stats import ttest_1samp

# load dataset

df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df1 = df[['age', 'sex']]
```

```python
df1.head()
```

```python
df1.describe()
```

```python
# check the age and campare with a known value of 45 years

ttest_1samp(df1['age'], 45)
```

```python
ttest_1samp(df1['age'], 45)
stat, p = ttest_1samp(df1['age'], 50)

print('stat=%.3f, p=%.3f' %(stat, p))

if p > 0.05:
    print('Probablity the same distribution')

else:
    print('Probablity different distribution')
        
```
