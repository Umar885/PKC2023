```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt     
 
```

```python
# normal distribution

def pdf(x):
    mean = np.mean(x)
    std = np.std(x)
    y_out = 1/(std*np.sqrt(2*np.pi))* np.exp( - (x-mean)**2/ (2*std**2))
    return y_out
# to genrate an array of x
x = np.arange(-2, 2, 0.1)
y = pdf(x)
# Plotting the normal curve / ball curve or gaussian distrubution

plt.style.use('seaborn')
plt.figure(figsize=(5,5))

plt.plot(x,y, color = 'blue', linestyle = 'dashed')

plt.scatter(x,y, marker='o', s=25, color='red')

```

```python
# import  a dataset

kashti = sns.load_dataset('titanic')
kashti.head()
```

```python
# taking subset

kashti = kashti[['sex', 'age', 'fare']]
kashti.head()
```

```python
# histogram test
sns.histplot(kashti[['age', 'fare']])
```

```python
sns.histplot(kashti['age'])
```

```python
# qq plot

from statsmodels.graphics.gofplots import qqplot
# qq normal plot
qqplot(kashti['age'])
plt.show()
```

```python
# shapirowilk test

from scipy.stats import shapiro

stat, p = shapiro(kashti['age'])
print('stat=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print('probably Gaussian or Normal Distribution')
else:
    print('Probably not Gaussian nor normal distribution')    
```

```python
from scipy.stats import normaltest

stat, p = normaltest(kashti['age'])
print('stat=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print('probably Gaussian or Normal Distribution')
else:
    print('Probably not Gaussian nor normal distribution') 
```
